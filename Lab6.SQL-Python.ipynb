```python
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine
import getpass
```

```python
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila' 
engine = create_engine(connection_string)
data = pd.read_sql_query('''SELECT title, length, 
RANK() OVER(ORDER BY length DESC) AS ranking
FROM sakila.film
WHERE length is not null or length <> 0
ORDER BY ranking;''', engine)

data.head(60)
```

```python
# another way of doing it, a bit simplier

query = '''SELECT title, length, 
RANK() OVER(ORDER BY length DESC) AS ranking
FROM sakila.film
WHERE length is not null or length <> 0
ORDER BY ranking;'''

data = pd.read_sql(query, engine)

data.head(60)
```
